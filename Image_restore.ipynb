```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
d=cv2.imread(r"/content/shizuka")
d=cv2.cvtColor(d,cv2.COLOR_BGR2GRAY)
plt.imshow(d,cmap='grey')
plt.axis('off')
```

```python

mean = 0
std_dev = 25
gaussian_noise = np.random.normal(mean, std_dev, d.shape)

noisy_image = d + gaussian_noise

noisy_image = np.clip(noisy_image, 0, 255)

noisy_image = noisy_image.astype(np.uint8)

denoised_image=cv2.GaussianBlur(noisy_image,(5,5),0)

plt.figure(figsize=(10, 5))
plt.subplot(3, 3, 1)
plt.title('Original Image')
plt.imshow(d, cmap='gray')
plt.axis('off')

plt.subplot(3, 3, 2)
plt.title('Noisy Image')
plt.imshow(noisy_image, cmap='gray')
plt.axis('off')

plt.subplot(3, 3, 3)
plt.title('Denoised Image')
plt.imshow(denoised_image, cmap='gray')
plt.axis('off')

plt.subplot(3,3,4)
plt.hist(d)
plt.title('Original Image Histogram')

plt.subplot(3,3,5)
plt.hist(noisy_image)
plt.title('Noisy Image Histogram')

plt.subplot(3,3,6)
plt.hist(denoised_image)
plt.title('Denoised Image Histogram')

plt.show()
```

```python
def add_salt_and_pepper_noise(d, prob):
 noisy_img = d.copy()
 noise = np.random.rand(*d.shape)
 noisy_img[noise < prob / 2] = 0
 noisy_img[noise > 1 - prob / 2] = 255
 return noisy_img

noisy_image = add_salt_and_pepper_noise(d, 0.5)
denoised_image = cv2.medianBlur(noisy_image, 5)

plt.figure(figsize=(10, 5))
plt.subplot(3, 3, 1)
plt.title('Original Image')
plt.imshow(d, cmap='gray')
plt.axis('off')

plt.subplot(3, 3, 2)
plt.title('Noisy Image')
plt.imshow(noisy_image, cmap='gray')
plt.axis('off')

plt.subplot(3, 3, 3)
plt.title('Denoised Image')
plt.imshow(denoised_image, cmap='gray')
plt.axis('off')

plt.subplot(3,3,4)
plt.hist(d)
plt.title('Original Image Histogram')

plt.subplot(3,3,5)
plt.hist(noisy_image)
plt.title('Noisy Image Histogram')

plt.subplot(3,3,6)
plt.hist(denoised_image)
plt.title('Denoised Image Histogram')

plt.show()
```

```python
def add_impulse_noise(d,prob):
  noisy_image=d.copy()
  noisy=np.random.rand(*d.shape)
  noisy_image[noisy<prob]=255
  return noisy_image

noised_image=add_impulse_noise(d,0.2)
denoised_image=cv2.medianBlur(noised_image,5)

plt.figure(figsize=(10,6))
plt.subplot(3,3,1)
plt.title('Original Image')
plt.imshow(d,cmap='grey')
plt.axis('off')

plt.subplot(3,3,2)
plt.title('Noisy Image')
plt.imshow(noised_image,cmap='grey')
plt.axis('off')

plt.subplot(3,3,3)
plt.title('Denoised Image')
plt.imshow(denoised_image,cmap='grey')
plt.axis('off')

plt.subplot(3,3,4)
plt.hist(d)
plt.title('Original Image Histogram')

plt.subplot(3,3,5)
plt.hist(noised_image)
plt.title('Noisy Image Histogram')

plt.subplot(3,3,6)
plt.hist(denoised_image)
plt.title('Denoised Image Histogram')

plt.show()
```

```python

def add_rayleigh_noise(d, scale=10):
 noise = np.random.rayleigh(scale, size=d.shape)
 noisy_image = d + noise
 noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)
 return noisy_image

def restore_rayleigh_noise(noisy_image, original_image, scale=10):
 restored_image = noisy_image - np.random.rayleigh(scale, size=noisy_image.shape)
 restored_image = np.clip(restored_image, 0, 255).astype(np.uint8)
 return restored_image

noisy_image = add_rayleigh_noise(d, scale=10)
denoised_image = restore_rayleigh_noise(noisy_image,d)
plt.figure(figsize=(10, 5))

plt.subplot(3,3,1)
plt.title('Original Image')
plt.imshow(d,cmap='grey')
plt.axis('off')

plt.subplot(3,3,2)
plt.title('Noisy Image')
plt.imshow(noised_image,cmap='grey')
plt.axis('off')

plt.subplot(3,3,3)
plt.title('Denoised Image')
plt.imshow(denoised_image,cmap='grey')
plt.axis('off')

plt.subplot(3,3,4)
plt.hist(d)
plt.title('Original Image Histogram')

plt.subplot(3,3,5)
plt.hist(noised_image)
plt.title('Noisy Image Histogram')

plt.subplot(3,3,6)
plt.hist(denoised_image)
plt.title('Denoised Image Histogram')

plt.show()
```

```python
def add_exponential_noise(d, lambda_val=0.02):
 scale = 1 / lambda_val
 noise = np.random.exponential(scale, d.shape).astype(np.uint8)
 noisy_image=d+noise
 noisy_img = np.clip(noisy_image, 0, 255)
 return noisy_img.astype(np.uint8)


lambda_value = 0.02
noisy_image = add_exponential_noise(d, lambda_value)
restored_image = cv2.medianBlur(noisy_image, 9)

plt.figure(figsize=(10, 5))

plt.subplot(3,3,1)
plt.title('Original Image')
plt.imshow(d,cmap='grey')
plt.axis('off')

plt.subplot(3,3,2)
plt.title('Noisy Image')
plt.imshow(noisy_image,cmap='grey')
plt.axis('off')

plt.subplot(3,3,3)
plt.title('Denoised Image')
plt.imshow(restored_image,cmap='grey')
plt.axis('off')

plt.subplot(3,3,4)
plt.hist(d)
plt.title('Original Image Histogram')

plt.subplot(3,3,5)
plt.hist(noisy_image)
plt.title('Noisy Image Histogram')

plt.subplot(3,3,6)
plt.hist(restored_image)
plt.title('Denoised Image Histogram')

plt.show()
```
