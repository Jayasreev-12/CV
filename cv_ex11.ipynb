```python
cap = cv2.VideoCapture(r"C:\Users\hanal\Desktop\girl with dog.mp4")
background = None
while True:
    r, f = cap.read()     
    if not r:
        break
    g = cv2.cvtColor(f, cv2.COLOR_BGR2GRAY)
    gf=cv2.resize(f,(500,500))   
    if background is None:
        background = gf.astype("float")
        continue
    cv2.accumulateWeighted(gf, background, 0.01)  # learning rate        
    diff_frame = cv2.absdiff(gf, cv2.convertScaleAbs(background))
    _, fg_mask = cv2.threshold(diff_frame, 25, 255, cv2.THRESH_BINARY)
    cv2.imshow('original',gf)
    cv2.imshow('Foreground Mask', fg_mask)     
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
c = cv2.VideoCapture(r"C:\Users\hanal\Downloads\8394092-uhd_2160_3840_24fps.mp4")
bg = cv2.createBackgroundSubtractorMOG2(history=500, varThreshold=100, detectShadows=False)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
ret, prev_frame = c.read()
prev_frame = cv2.resize(prev_frame, (500, 500))  # Resize first frame
prev_frame = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)
while True:
    ret, frame = c.read()
    if not ret:
        break
    frame_resized = cv2.resize(frame, (500, 500))  # Resize every frame
    gray = cv2.cvtColor(frame_resized, cv2.COLOR_BGR2GRAY)
    diff = cv2.absdiff(prev_frame, gray)
    _, diff_thresh = cv2.threshold(diff, 50, 255, cv2.THRESH_BINARY)
    if np.sum(diff_thresh) > 5_000_000:  
        print("Scene Change Detected ")
        bg = cv2.createBackgroundSubtractorMOG2(history=500, varThreshold=100, detectShadows=False)
    prev_frame = gray.copy()  
    fg = bg.apply(frame_resized)
    fg = cv2.morphologyEx(fg, cv2.MORPH_OPEN, kernel)
    fg = cv2.morphologyEx(fg, cv2.MORPH_CLOSE, kernel)
    cv2.imshow('Original', frame_resized)
    cv2.imshow('Motion Detection', fg)
    if cv2.waitKey(30) & 0xFF == 27:  # Press 'Esc' to exit
        break
c.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
def detect(v):
    cap = cv2.VideoCapture(v)
    background = None
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        gray_frame=cv2.resize( gray_frame,(500,500))
        if background is None:
            background = gray_frame.astype("float")
            continue
        cv2.accumulateWeighted(gray_frame, background, 0.01)
        diff_frame = cv2.absdiff(gray_frame, cv2.convertScaleAbs(background))
        _, fg_mask = cv2.threshold(diff_frame, 25, 255, cv2.THRESH_BINARY)
        changed_pixels = cv2.countNonZero(fg_mask)
        if changed_pixels > 50000:
            print("Scene change detected!")
        cv2.imshow('original',gray_frame)
        cv2.imshow('Foreground Mask', fg_mask)
        k=cv2.waitKey(30) 
        if k == 30:
            break
    cap.release()
    cv2.destroyAllWindows()
detect (r"C:\Users\hanal\Desktop\girl with dog.mp4")
```
