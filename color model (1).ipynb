```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
d=cv2.imread(r"C:\Users\hanal\Downloads\image5.jpg")
```

```python
plt.imshow(d)
```

```python
RGB=cv2.cvtColor(d,cv2.COLOR_BGR2RGB)
```

```python
r,g,b=RGB[:,:,0],RGB[:,:,1],RGB[:,:,2]
```

```python
c=1-r/225
m=1-g/225
y=1-b/225
```

```python
i=(np.dstack((c,m,y))*255).astype(np.uint8)

hsv=cv2.cvtColor(d,cv2.COLOR_BGR2HSV)
```

```python
r1=r/225
g1=g/225
b1=b/225
y=0.299*r1 + 0.587*g1 + 0.114*b1 
ii=0.596*r1 - 0.274*g1 - 0.321*b1
q=0.211*r1 - 0.522*g1 + 0.311*b1
yiq=(np.dstack((y,ii,q))*255).astype(np.uint8)
```

```python
plt.figure(figsize=(10,5 ))
plt.subplot(2,2,1)
plt.imshow(d)
plt.title('ORIGINAL')
plt.axis('off')
plt.subplot(2,2,2)
plt.imshow(i)
plt.title('CMY')
plt.axis('off')
plt.subplot(2,2,3)
plt.imshow(hsv)
plt.title('HSV')
plt.axis('off')
plt.subplot(2,2,4)

plt.title('YIQ')
plt.imshow(yiq)
plt.axis('off')
```

```python
RGB1=cv2.cvtColor(d,cv2.COLOR_BGR2RGB)
RGB2=cv2.cvtColor(d,cv2.COLOR_BGR2RGB)
RGB3=cv2.cvtColor(d,cv2.COLOR_BGR2RGB)
```

```python
RGB1[:,:,1]=0
RGB1[:,:,2]=0
RGB2[:,:,0]=0
RGB2[:,:,2]=0
RGB3[:,:,0]=0
RGB3[:,:,1]=0
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1,3,1)
plt.title('red')
plt.axis('off')
plt.imshow(RGB1)
plt.subplot(1,3,2)
plt.title('green')
plt.axis('off')
plt.imshow(RGB2)
plt.subplot(1,3,3)
plt.title('blue')
plt.axis('off')
plt.imshow(RGB3)
```

```python
h = cv2.cvtColor(d, cv2.COLOR_BGR2HSV)
H, S, V = cv2.split(h)
plt.figure(figsize=(12, 4))

plt.subplot(1,3,1)
plt.imshow(H, cmap='hsv')
plt.title('Hue Channel')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(S, cmap='gray')
plt.title('Saturation Channel')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(V, cmap='gray')
plt.title('Value Channel')
plt.axis('off')

plt.show()
```

```python
h = cv2.cvtColor(d, cv2.COLOR_BGR2HSV)
h1 = cv2.cvtColor(d, cv2.COLOR_BGR2HSV)
h2 = cv2.cvtColor(d, cv2.COLOR_BGR2HSV)
h3 = cv2.cvtColor(d, cv2.COLOR_BGR2HSV)
```

```python
h1[:,:,1]=100
h2[:,:,2]=240
h3[:,:,0]=180
```

```python
plt.subplot(2,2,1)
plt.title('HSV')
plt.imshow(h)
plt.axis('off')

plt.subplot(2,2,2)
plt.title('SATURATION ALTERED')
plt.imshow(h1)
plt.axis('off')

plt.subplot(2,2,3)
plt.title('VALUE ALTERED')
plt.imshow(h2)
plt.axis('off')

plt.subplot(2,2,4)
plt.title('HUE ALTERED')
plt.imshow(h3)
plt.axis('off')
```
