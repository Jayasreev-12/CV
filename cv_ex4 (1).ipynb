```python
import cv2
import matplotlib.pyplot as plt
d=cv2.imread(r"C:\Users\hanal\Downloads\unnamed.jpg")
d=cv2.cvtColor(d,cv2.COLOR_BGR2GRAY)
plt.imshow(d,cmap='grey')
plt.axis('off')
```

```python
# b=cv2.GaussianBlur(d,(201,201),0)
import numpy as np
r,c=d.shape
b=np.zeros_like(d)
k= np.array([[1, 2, 1], [2, 4,2], [1, 2, 1]],dtype=np.uint8)/16

for i in range(1, r - 1):
    for j in range(1, c - 1): 
        re = d[i-1:i+2, j-1:j+2]
        n = np.sum(re * k)
        b[i, j] = int(n)

plt.subplot(221)
plt.imshow(d,cmap='grey')
plt.axis('off')
plt.subplot(222)
plt.axis('off')
plt.imshow(b,cmap='grey')
plt.subplot(223)
plt.hist(d)
plt.subplot(224)
plt.hist(b)
```

```python
# val,t= cv2.threshold(d, 100, 255, cv2.THRESH_BINARY)

import numpy as np
def thr(y, n):
    # y = np.array(y, dtype=np.uint8)
    r,c = y.shape
   
    for i in range(r):
        for j in range(c):
            if y[i, j] > n:
                y[i, j] = 0
            else:
                y[i, j] = 255
    return y

t=thr(d,100)
d=cv2.imread(r"C:\Users\hanal\Downloads\unnamed.jpg")
d=cv2.cvtColor(d,cv2.COLOR_BGR2GRAY)
plt.subplot(221)
plt.imshow(d,cmap='grey')

plt.axis('off')
plt.subplot(222)
plt.axis('off')
plt.imshow(t,cmap='grey')
plt.subplot(223)
plt.hist(d)
plt.subplot(224)
plt.hist(t)
```
